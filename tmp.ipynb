```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

def load_model_and_tokenizer(model_path):
    tokenizer = AutoTokenizer.from_pretrained(model_path, trust_remote_code=True)
    model = AutoModelForCausalLM.from_pretrained(
        model_path, 
        torch_dtype=torch.float16, 
        trust_remote_code=True,
        device_map="auto"  # 这将自动处理模型在可用设备上的分配
    )
    
    if tokenizer.pad_token is None:
        tokenizer.pad_token = tokenizer.eos_token
    tokenizer.padding_side = "right"
    
    return model, tokenizer

# 在主代码中使用
model_path = "/root/models/Shanghai_AI_Laboratory/internlm2-chat-1_8b"
model, tokenizer = load_model_and_tokenizer(model_path)
```

```python
model = model.eval()
response, history = model.chat(tokenizer, "hello", history=[])
print(response)
# Hello! How can I help you today?
response, history = model.chat(tokenizer, "please provide three suggestions about time management", history=history)
print(response)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 生成模拟数据
np.random.seed(0)
study_hours = np.random.uniform(1, 10, 1000)  # 学习时间：1到10小时
scores = study_hours * 10 + np.random.normal(0, 5, 1000)  # 考试成绩：学习时间的10倍，加上随机噪声

# 创建热力图
plt.figure(figsize=(10, 6))
sns.histplot(
    x=study_hours, 
    y=scores, 
    bins=30,  # 网格的数量
    pthresh=0.1, 
    cmap="YlGnBu"
)
plt.xlabel('Study Hours')
plt.ylabel('Scores')
plt.title('Heatmap of Study Hours vs Scores')
plt.show()
```
